```python
# Importing the dataset 
sales = pd.read_csv("C:/Users/Juliana Hubacova/OneDrive - McGill University/Desktop/Github/Portfolio/Project_1/Data/amazon.csv")
# Exploring the dataset
sales.head()
```

All variables are currently of type object, which limits numerical operations and efficient processing. To prepare the data for analysis and modeling, we will perform the following steps:
- *Categorical Variables*: We have identified the columns that are inherently categorical, such as `product_id`, `category`, and `user_name`.
These columns will be converted to the category type to optimize memory usage and processing efficiency.
- *Numerical Variables*: Columns such as `discounted_price`, `actual_price`, `discount_percentage`, `rating`, and `rating_count` are numerical by nature but are currently stored as object. These columns will be cleaned and converted to numerical types (float) for accurate analysis.
    - For prices (`discounted_price`, `actual_price`): Remove the currency symbol (₹) and commas for proper numerical conversion.
    - For discount percentages (`discount_percentage`): Remove the percentage symbol (%) to retain only the numerical value.
    - For rating counts (`rating_count`): Remove commas to convert the column into numeric format.
    - Invalid Data Handling: In the `rating` column, we identified invalid values such as the symbol |, which does not represent a numerical rating. We will remove rows containing such invalid entries.

```python
# Defining categorical variables
categorical_variables = ['product_id', 'product_name', 'category', 'about_product', 'user_id', 'user_name', 'review_id', 'review_title', 'review_content', 'img_link', 'product_link']

# Defining numerical variables
numerical_variables = ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']

# For discounted_price and actual_price, I will remove the currency sign and the comma
sales['discounted_price'] = sales['discounted_price'].str.replace('₹', '', regex=True).str.replace(',', '')
sales['actual_price'] = sales['actual_price'].str.replace('₹', '', regex=True).str.replace(',', '')

# For discounted_percentage I will remove the % sign
sales['discount_percentage'] = sales['discount_percentage'].str.replace('%', '', regex=True)

# For the rating_count I will remove the comma
sales['rating_count'] = sales['rating_count'].str.replace(',', '')

# For the rating column
sales['rating'].value_counts()

# Deleting the observation where the rating is '|'
sales = sales[sales['rating'] != '|']

# Converting to float
for col in numerical_variables:
    sales[col] = sales[col].astype(float)
```

#### Exploring the Distribution of `Ratings`
Checking the skewness of the rating variable is essential because it helps us understand the distribution of our target variable. Many machine learning models, such as linear regression, assume that the target variable follows a normal distribution for better predictions and unbiased estimations. If the rating distribution is heavily skewed, it can impact the model's performance by introducing bias or inefficiency in optimization. Understanding the skewness also helps determine whether a transformation (e.g., log or square root) is needed to normalize the data, especially for models sensitive to outliers or asymmetrical distributions.

```python
# Histogramme de la colonne 'rating'
plt.figure(figsize=(8, 5))
plt.hist(sales['rating'].dropna(), bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Ratings', fontsize=16)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate skewness of the 'rating' variable
skewness = sales['rating'].skew()
print(f"Skewness of ratings: {skewness}")
```

The `rating` variable has a left-skewed distribution, with a skewness value of -1.24. This means that most ratings fall in the higher range (4 to 4.5), while a smaller number of lower ratings (below 3.5) create a tail on the left side. This suggests that most products are rated positively, which might indicate good overall customer satisfaction. However, this skewness could affect machine learning models that assume the target variable is normally distributed. Recognizing this skewness is important to decide if transformations (like squaring) are necessary or if non-normality-tolerant models should be used.

To handle left-skewed data we can use the following techniques: square transformation, exponential transformation, or inverse transformation. I tried all those transformation and the one that had the lowest skewness in absolute value was 0.27 for the exponential transformation, compared to -0.75 for the square transformation and 3.22 for the inverse transformation.

```python
sales['rating_transformed_exp'] = np.exp(sales['rating'])
# Histogram 'rating'
plt.figure(figsize=(8, 5))
plt.hist(sales['rating_transformed_exp'].dropna(), bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Ratings', fontsize=16)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate skewness of the 'rating' variable
skewness = sales['rating_transformed_exp'].skew()
print(f"Skewness of ratings: {skewness}")
```

Target Variable Selection for Machine Learning Algorithms
The `rating` variable will be used as the target variable for machine learning algorithms that are *not sensitive* to skewness, such as tree-based models (e.g., Random Forest, Gradient Boosting) and Naive Bayes. These algorithms are robust to the distribution of the target variable and do not require normalization. 

On the other hand, the transformed rating variable `rating_transformed_exp`, created using an exponential transformation, will be used for algorithms that are *sensitive* to skewness, such as Linear Regression, Logistic Regression, and Neural Networks. This approach ensures that the models are provided with a target variable suitable for their specific assumptions, maximizing performance and accuracy.

#### Handling Missing Values 
Dealing with missing values is essential to ensure accuracy and reliability of machine learning models and analyses. Missing data can lead to inaccurate summaries and poor model performance, especially if a lot of important information is missing. If we don't handle them properly, it could introduce errors or bias into our analysis. By taking care of missing values - either by filling them in, removing them, or using other techniques - we make sure our data is clean and our results are accurate and trustworthy.

```python
# Checking for missing values in the dataset
missing_values = sales.isnull().sum()

# Display columns with missing values
print(missing_values[missing_values > 0])

# Visualize missing data
plt.figure(figsize=(10, 6))
msno.matrix(sales, color=(0.2, 0.4, 0.6))
plt.title("Missing Values Visualization", fontsize=22)
plt.show()
```

From the visualization, we can see that the dataset has minimal missing values, specifically in the `rating_count` column, while all other columns are fully populated. This indicates that the data is mostly complete and well-suited for analysis. The small number of missing values in `rating_count` is unlikely to significantly impact the overall analysis but should still be addressed. I will further inspect the observations with missing values in `rating_count` to understand their context before deciding whether to impute or remove them.

```python
# Checking the observation with the missing values
sales[sales['rating_count'].isnull()]
```

Upon inspection, there are two observations in the dataset with missing values in the `rating_count` column. Both observations have valid `rating` values (3.0 and 5.0), but their corresponding `rating_count` is missing. This could indicate that while a rating was provided, the number of users who rated the product is unavailable. To address this, I will impute the missing values in `rating_count` using the median of the column, as the median is less sensitive to outliers and will provide a more robust estimate compared to the mean.

```python
# Fill with the median
sales['rating_count'] = sales.rating_count.fillna(value=sales['rating_count'].median())

# Checking if there still are missing values
sales.isnull().sum()
```

#### Outliers

Outlier detection and removal are essential steps in data preprocessing because outliers can heavily influence the accuracy of analysis and machine learning models. Outliers can skew statistical metrics like the mean, distort distributions, and lead to biased or unreliable model predictions. To address this, I will use boxplots to visually identify outliers in the dataset. This approach allows for a clear understanding of the range of values in each variable and helps determine whether the outliers represent errors, anomalies, or meaningful data points. Identifying and handling outliers ensures the dataset is clean and more suitable for analysis.

```python
# Looping through each numeric column and creating boxplots
for col in numerical_variables:
    plt.figure(figsize=(8, 4))
    plt.boxplot(sales[col], vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.show()
```

On the other hand, outliers in `rating` and `rating_count` likely reflect rare low ratings or highly popular products, which are meaningful and should be retained. To balance accuracy and information preservation, I will carefully assess the outliers, removing only those in `discounted_price` and `actual_price` if they appear to be errors or overly distort the analysis. For `rating` and `rating_count`, the outliers will be retained to maintain the integrity of customer trends and product performance insights.

#### For Categorical Variables

```python
# Summarize categories of product
sales['category'].value_counts()
```

The `category` column in the dataset has a hierarchical structure, with multiple levels of categories separated by `|` (e.g., `Electronics|HomeTheater|Televisions`). While this structure provides detailed information about subcategories, it may be too granular for the purpose of analysis and modeling. To simplify the dataset and focus on higher-level trends, I will group the data by broad categories (e.g., `Electronics`, `Home&Kitchen`). This approach allows for a clearer understanding of product trends across main categories while reducing complexity. By working with broad categories, we can maintain meaningful insights without being overwhelmed by excessive detail.

```python
# Extracting the broader category
sales['broad_category'] = sales['category'].str.split('|').str[0]

# Summarize categories of product
sales['broad_category'].value_counts()
```

```python
# Horizontal bar plot
sales['broad_category'].value_counts().plot(kind='barh', color='skyblue', edgecolor='black', figsize=(10, 6))
plt.title('Distribution of Broad Categories', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Broad Category', fontsize=12)
plt.show()
```

The `Electronics` category has the highest count, with over 500 products, making it the largest and most dominant category in the dataset. Along with `Computers&Accessories` and `Home&Kitchen`, which each have over 400 products, these three categories represent the majority of the dataset and are likely to play a significant role in any trends or patterns observed. On the other hand, smaller categories such as `OfficeProducts`, `MusicalInstruments`, and `HomeImprovement` are represented by very few products, while categories like `Toys&Games`, `Car&Motorbike`, and `Health&PersonalCare` have only one product each. To simplify the analysis, these smaller categories will be grouped into an "Other" category, ensuring the focus remains on the most representative product categories.

```python
# Categories with fewer than 10 products will be grouped as "Other"
threshold = 50

# Grouping smaller categories into "Other"
sales['broad_category_grouped'] = sales['broad_category'].apply(
    lambda x: x if sales['broad_category'].value_counts()[x] >= threshold else 'Other'
)

# Verify the new grouping
print(sales['broad_category_grouped'].value_counts())
```

```python
# Pie chart
sales['broad_category_grouped'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), colors=plt.cm.Paired.colors)
plt.title('Proportion of Broad Categories', fontsize=16)
plt.ylabel('')  # Remove the y-axis label for better aesthetics
plt.show()
```
